```python
import matplotlib.pyplot as plt
```

<h1 style="color:black;">Creating our first chart</h1>


```python
#data

x = [1,2,3,4,5,6]
y = [10,20,25,30,35, 40]


plt.plot(x,y)
plt.show


```

<h1 style:"color:black;">Customize this chart</h1>

```python
plt.figure(figsize=(4,3))

plt.plot(x, y, color='green', marker='v', linestyle='dashed', linewidth=2, markersize=12)

plt.title("Data title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

```

<h1 style="color:black;">Multiline chart</h1>

```python
x = [1,2,3,4,5,6]
y1 = [10,20,25,30,35, 40]
y2 = [20, 30, 40, 50, 60, 70] 



plt.plot(x, y1, label = "Sales 2025")
plt.plot(x, y2, label = "Sales 2026")


plt.title("yoy sales")
plt.xlabel("Months")
plt.ylabel("Sales")


plt.legend()
plt.show()

```

<h1 style="color:black;">BAR Chart</h1>

```python
x = ["HR","Sales","Marketing","finance"]
y = [20,30,35, 40]

plt.bar(x,y)

plt.title("Departments Employee Count")
plt.show()
```

<h1 style="color:black;">Histogram</h1>

```python
import random

data = [random.randint(1,20) for _ in range(300)]



plt.hist(data, bins = 15)
plt.title("title")
plt.show()
```

<h1 style="color:black;">PIE CHART </h1>  

100 - 20, 40, 15, 25|| whole -> category define

```python
categories = ["HR","Sales","Marketing","finance", "Business"]
n = [20, 26, 34, 7, 13]



plt.pie(n, labels=categories, autopct = '%1.1f%%', startangle = 90)

plt.title("department structure")
plt.show()


```

<h1 style="color:black;">  Scatter plot </h1>

```python

# Cluster 1
x1 = [2, 3, 2.5, 3.5, 4, 3.2, 2.8]
y1 = [8, 8.5, 7.8, 9, 8.7, 8.2, 7.9]

# Cluster 2
x2 = [8, 9, 8.5, 9.5, 10, 9.2, 8.8]
y2 = [2, 2.5, 3, 2.8, 3.2, 2.1, 2.7]

# Cluster 3
x3 = [6, 6.5, 7, 7.5, 6.8, 7.2, 6.3]
y3 = [6, 6.5, 7, 6.8, 7.2, 6.3, 6.7]






plt.scatter(x1, y1, label= "Cluster 1")
plt.scatter(x2, y2, label= "Cluster 2")
plt.scatter(x3, y3, label= "Cluster 3")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()

plt.show
```

<h1 style="color:black;"> Subplot </h1>

```python
x = [1,2,3,4,5]


y1 = [2,4,6,8,10]
y2 = [1,4,9,16,25]
y3 = [5,7,6,9,11]
y4 = [10,8,7,4,3]



fig,axs = plt.subplots(2,2, figsize=(10,8))


# plt.subplot(1,2,1)
# plt.bar(x, y1)
# plt.xlabel =" x axis 1"
# plt.ylabel = "y axis 1"


# plt.subplot(1,2,2)
# plt.scatter(x, y2)







#subplot 1
axs[0,0].plot(x,y1)
axs[0,0].set_title("Linear Growth")


#subplot 2
axs[0,1].plot(x,y2)
axs[0,1].set_title("Quadratic Growth")


#subplot 3
axs[1,0].scatter(x,y3)
axs[1,0].set_title("Scatter plot")

#subplot 4
axs[1,1].bar(x,y4)
axs[1,1].set_title("Bar plot")


plt.show()
```

<h1 style="color:blue";> Vector dot product representation </h1>

```python
import numpy as np
```

```python
person = np.array([25, 60000, 3])   #a person's features


weights = np.array([0.1, 0.0001, 2.01])
```

```python
score = np.dot(person, weights)
print(score) #prediction score
```

```python
user_1 = np.array([1,0,1,1])    #liked A, C, D 
user_2 = np.array([1,1,1,0])    # liked A, B, D


simlilarity = np.dot(user_1, user_2)
print(similarity)
```
